```python
import tensorflow as tf
import numpy as np
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,Activation,Dropout,Conv2D,MaxPooling2D
```

```python
import zipfile
import os
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
with zipfile.ZipFile('/content/training_set.zip','r') as zip_ref:
  zip_ref.extractall('/content/data')

  data= ImageDataGenerator(rescale=1./255)

  x_train = data.flow_from_directory(
      '/content/data/training_set',
       target_size=(64,64),
       batch_size=32,
       class_mode='binary')


  print(x_train.class_indices)
```

```python
with zipfile.ZipFile('/content/test_set.zip','r') as zip_ref:
  zip_ref.extractall('/content/data')

  data= ImageDataGenerator(rescale=1./255)

  x_test = data.flow_from_directory(
      '/content/data/test_set',
       target_size=(64,64),
       batch_size=32,
       class_mode='binary')


  print(x_train.class_indices)
```

```python
model = Sequential()

model.add(Conv2D(32,(3,3),input_shape=(64,64,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3),input_shape=(64,64,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dense(64))

model.add(Dense(1))
model.add(Dropout(0.5))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(x_train,epochs=10,batch_size=32,validation_data=x_test)
model.summary()
```

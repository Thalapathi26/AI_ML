```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/personality_datasert.csv")
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df["Personality"]=le.fit_transform(df["Personality"])
df['Drained_after_socializing']=le.fit_transform(df['Drained_after_socializing'])
df["Stage_fear"]=le.fit_transform(df["Stage_fear"])
df
```

```python
x=df.drop("Personality",axis=1)
y=df["Personality"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report

accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python

from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Re-initializing LabelEncoders with known classes for consistency
# These need to be consistent with how they were trained

le_stage_fear = LabelEncoder()
le_stage_fear.fit(['No', 'Yes'])

le_drained = LabelEncoder()
le_drained.fit(['No', 'Yes'])

le_personality = LabelEncoder()
le_personality.fit(['Extrovert', 'Introvert']) # Assuming 'Extrovert' was 0 and 'Introvert' was 1 based on previous output

def predict_personality(time_spent_alone, stage_fear, social_event_attendance, going_outside, drained_after_socializing, friends_circle_size, post_frequency):
    # Encode categorical inputs
    encoded_stage_fear = le_stage_fear.transform([stage_fear])[0]
    encoded_drained = le_drained.transform([drained_after_socializing])[0]

    # Create a DataFrame for the new input
    input_data = pd.DataFrame([[time_spent_alone, encoded_stage_fear, social_event_attendance, going_outside, encoded_drained, friends_circle_size, post_frequency]],
                              columns=x.columns)

    # Make prediction
    prediction_encoded = dt.predict(input_data)[0]

    # Inverse transform the prediction to get the original label
    predicted_personality = le_personality.inverse_transform([prediction_encoded])[0]

    return predicted_personality

# --- Get user inputs ---
print("Please enter the following details to predict personality:")

try:
    time_spent_alone = float(input("Time spent alone (e.g., 5.0): "))
    stage_fear = input("Stage fear (Yes/No): ").capitalize()
    social_event_attendance = float(input("Social event attendance (e.g., 7.0): "))
    going_outside = float(input("Going outside (e.g., 8.0): "))
    drained_after_socializing = input("Drained after socializing (Yes/No): ").capitalize()
    friends_circle_size = float(input("Friends circle size (e.g., 10.0): "))
    post_frequency = float(input("Post frequency (e.g., 6.0): "))

    # Validate categorical inputs
    if stage_fear not in ['Yes', 'No']:
        raise ValueError("Stage fear must be 'Yes' or 'No'.")
    if drained_after_socializing not in ['Yes', 'No']:
        raise ValueError("Drained after socializing must be 'Yes' or 'No'.")

    # Get prediction
    result = predict_personality(time_spent_alone, stage_fear, social_event_attendance, going_outside, drained_after_socializing, friends_circle_size, post_frequency)

    print(f"\nBased on your inputs, the predicted personality is: {result}")

except ValueError as e:
    print(f"Invalid input: {e}. Please ensure all numerical inputs are valid numbers and categorical inputs are 'Yes' or 'No'.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

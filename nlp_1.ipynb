```python
import pandas as pd
import numpy as np
import re
import nltk

from nltk.corpus import stopwords #remove and,or,is
from nltk.stem import WordNetLemmatizer #change root form
from sklearn.feature_extraction.text import TfidfVectorizer # change numerical method
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
df=pd.read_csv('/content/Restaurant_Reviews.tsv', delimiter='\t')
```

```python
df
```

```python
stopwords=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
def preprocess(text):
  text=re.sub('[^a-zA-Z]',' ',text)
  text=text.lower()
  words=text.split()
  words=[lemmatizer.lemmatize(word) for word in words if word not in stopwords]
  return" ".join(words)

df['Review']=df['Review'].apply(preprocess)
df
```

```python
vectorizer=TfidfVectorizer()
x=vectorizer.fit_transform(df['Review'])
y=df['Liked']
```

```python
df['Liked'].value_counts()
```

```python
x=(df['Review'])
y=(df['Liked'])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, SimpleRNN
```

```python
print(x_train)
```

```python
#x_train = x_train.toarray()
#x_test = x_test.toarray()
```

```python
model=Sequential()
model.add(Embedding(input_dim=10000,output_dim=32))
model.add(SimpleRNN(512,activation='tanh'))

model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
from sklearn.utils import validation
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
lstm=model.fit(x_train, y_train, epochs=15, batch_size=32, validation_data=(x_test, y_test))
```

```python
prediction=model.predict(x_test)
prediction
```
